```python
import requests
from bs4 import BeautifulSoup as bs
import requests
from urllib.request import urlopen as uReq
import pandas as pd
```

```python
Product_name = []
Prices = []
Description = []
Reviews = []
```

```python
for i in range(2,12):
    
                url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
                r = requests.get(url)
                urlClient = uReq(url)
                soup = bs(urlClient,'html.parser')
                np = soup.find("a", class_="_1LKTO3").get("href")
                cnp = "https://www.flipkart.com"+np
                print(cnp)
                box = soup.find("div", class_="_1YokD2 _3Mn1Gg")
                names = box.find_all("div", class_ = "_4rR01T")
                for i in names:
                    name = i.text
                    Product_name.append(name)
                print(Product_name)
                pr = soup.find_all("div", class_ = "_30jeq3 _1_WHN1")
                for i in pr:
                    price = i.text
                    Prices.append(price)
                print(Prices)
                desc = soup.find_all("ul", class_ = "_1xgFaf")
                for i in desc:
                    d = i.text
                    Description.append(d)
                print(Description)
                rv = box.find_all("div", class_ = "_3LWZlK")
                for i in rv:
                    r = i.text
                    Reviews.append(r)
                print(Reviews)
                
```

```python
a ={"Product Name":Product_name, "Price":Prices, "Description":Description, "Review":Reviews}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```

```python
df.to_csv("Flipkart Web-Scrapping.csv")
```
